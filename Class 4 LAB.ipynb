```python
#if you want to do aggregations on data 

#what is the avg # of visitors per restaurant? 

df.groupby('id')
```

```python
df.groupby('id').mean()
```

```python
#Let's expand on this 

#What if I want to group by multiple columns? 

#Example: 

df.groupby(['id', 'day_of_week']).mean()['visitors']
```

```python
#all numeric columns 

df.groupby(['id', 'day_of_week']).mean()
```

```python
df.groupby(['id', 'day_of_week']).mean().index
```

```python
df.groupby(['id', 'day_of_week']).mean().sort_index(level = 1) #sort by day of week
```

```python
df.groupby(['id', 'day_of_week']).mean().sort_index(level = [0,1]) #sort by id and day of week
```

```python
df.groupby(['id', 'day_of_week']).mean().reset_index()
```

```python
df.groupby(['id', 'day_of_week']).mean()
```

```python
df.groupby(['id', 'day_of_week']).sum()
```

```python
#additional grouping method 

#x refers to each unique groupings 
df.groupby(['id', 'day_of_week']).apply(lambda x: x.isnull().sum())
```

```python
#Statistics for each individual column 
df.groupby('id').agg([np.mean, np.max, np.min, np.median])
```
